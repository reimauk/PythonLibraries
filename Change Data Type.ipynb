```python
import pandas as pd
```

```python
def changeDatatype(students):
    students['grade'] = students['grade'].astype('int')
    return students
```

```python
# Пример DataFrame
data = {
    'student_id': [1, 2],
    'name': ['Ava', 'Kate'],
    'age': [6, 15],
    'grade': [73.0, 87.0]
}

students = pd.DataFrame(data)
print(students)
print()

# Вызываем функцию для изменения типа данных столбца 'grade'
result = changeDatatype(students)
print(result)
```
