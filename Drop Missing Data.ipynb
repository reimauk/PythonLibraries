```python
import pandas as pd
```

```python
def dropMissingData(students):
    students = students.dropna(subset=['name'])
    return students
```

```python
# Пример DataFrame
data = {
    'student_id': [32, 217, 779, 849],
    'name': ['Piper', None, 'Georgia', 'Willow'],
    'age': [5, 19, 20, 14]
}

students = pd.DataFrame(data)
print(students)
print()
# Вызываем функцию для удаления строк с пропущенными значениями в колонке 'name'
result = dropMissingData(students)
print(result)
```
