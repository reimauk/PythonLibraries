```python
import pandas as pd
```

```python
def renameColumns(students):
    students = students.rename(columns={
        'id':'student_id',
        'first': 'first_name',
        'last': 'last_name',
        'age': 'age_in_years'
    })
    return students
```

```python
# Пример DataFrame
data = {
    'id': [1, 2, 3, 4, 5],
    'first': ['Mason', 'Ava', 'Taylor', 'Georgia', 'Thomas'],
    'last': ['King', 'Wright', 'Hall', 'Thompson', 'Moore'],
    'age': [6, 7, 16, 18, 10]
}

students = pd.DataFrame(data)

# Вызываем функцию для переименования столбцов
result = renameColumns(students)
print(result)
```
