```python
import pandas as pd
```

```python
def dropDuplicateEmails(customers):
    customers = customers.drop_duplicates(subset='email', keep='first')
    return customers
```

```python
# Пример DataFrame
data = {
    'customer_id': [1, 2, 3, 4, 5, 6],
    'name': ['Ella', 'David', 'Zachary', 'Alice', 'Finn', 'Violet'],
    'email': ['emily@example.com', 'michael@example.com', 'sarah@example.com', 'john@example.com', 'john@example.com', 'alice@example.com']
}

customers = pd.DataFrame(data)
print(customers)
print()

# Вызываем функцию для удаления дубликатов по колонке 'email'
result = dropDuplicateEmails(customers)
print(result)
```

```python
import pandas as pd
```

```python
def dropDuplicates(customers):
    customers = customers.drop_duplicates(subset=['email'], keep='first')
    return customers
```

```python
# Пример DataFrame
data = {
    'customer_id': [1, 2, 3, 4, 5, 6, 1],
    'name': ['Ella', 'David', 'Zachary', 'Alice', 'Finn', 'Violet', 'Alice'],
    'email': ['emily@example.com', 'michael@example.com', 'sarah@example.com', 'john@example.com', 'john@example.com', 'alice@example.com', 'alice@example.com']
}

customers = pd.DataFrame(data)
print(customers)
print()

# Вызываем функцию для удаления дубликатов по колонке 'email'
result = dropDuplicates(customers)
print(result)
```
